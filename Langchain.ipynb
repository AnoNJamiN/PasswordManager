```python
# Simplified LangChain Prompt Example

from langchain_core.prompts import ChatPromptTemplate


messages_recipe = [
    ("system", "You are a helpful expert on {topic}."),
    ("user", "Please tell me {number} interesting facts about {topic}.")
]

prompt_template = ChatPromptTemplate(messages=messages_recipe)


print("Let's build a prompt for an AI!")
input_topic = input("What topic do you want to ask about? ")
input_number = input(f"How many facts about {input_topic} do you want? ")


final_prompt = prompt_template.invoke({
    "topic": input_topic,
    "number": input_number
})

print("\n======================================")
print("✅ Here is your completed AI prompt:")
print("======================================")

# The 'final_prompt' object contains a list of messages. Let's look at them.
for message in final_prompt.to_messages():
    # 'message.type' will be 'system' or 'human'
    # 'message.content' is the actual text
    print(f"Role: {message.type.capitalize()}\nText: {message.content}\n")
```

```python
#no memory for multiturn chatbot
from langchain_groq import ChatGroq
#from dotenv import load_dotenv
import os

from langchain_core.messages import SystemMessage, HumanMessage, AIMessage

#load_dotenv()
groq_api_key = "gsk_Hidden"#os.getenv("GROQ_API_KEY")

llm = ChatGroq(model="llama-3.3-70b-versatile", api_key=groq_api_key)

ans=llm.invoke(final_prompt)

print("Ans:",ans)
print ("Type of Ans:",type(ans))
print(ans.content)
```

```python
#no memory for multiturn chatbot
from langchain_groq import ChatGroq
#from dotenv import load_dotenv
import os

from langchain_core.messages import SystemMessage, HumanMessage, AIMessage

#load_dotenv()

llm = ChatGroq(model="llama-3.3-70b-versatile", api_key=groq_api_key)

messages=[
    SystemMessage(content="You are a good assistant in Mathematics"),
    HumanMessage(content="What is 5 multiplied by 9?")

]

ans1=llm.invoke(messages)
print("Ans 1:")
print(ans1.content)

ans2=llm.invoke("What if one of the numbers is negative?")
print("Ans 2:")
print(ans2.content)

messages=[
    SystemMessage(content="You are a good assistant in Mathematics"),
    HumanMessage(content="What is 5 multiplied by 9?"),
    AIMessage(content=ans1.content),
    HumanMessage(content="What if one of the numbers is negative?")
]
ans3=llm.invoke(messages)
print("Ans 3:")
print(ans3.content)

messages=[
    ("system","You are a good assistant in Mathematics"),
    ("human","What is 5 multiplied by 9?"),
    ("ai",ans1.content),
    ("human","What if one of the numbers is negative?")
]

ans4=llm.invoke(messages)
print("Ans 4:")
print(ans4.content)
```

```python
#multiturn support
from langchain_groq import ChatGroq
#from dotenv import load_dotenv
import os

from langchain_core.messages import SystemMessage, HumanMessage, AIMessage


llm = ChatGroq(model="llama-3.3-70b-versatile", api_key=groq_api_key)

chat_history=[]
chat_history.append(SystemMessage(content="You are a good assistant in Mathematics"))

while True:
    human_message=input("Your Query:")
    if human_message.lower()=="exit":
        break
    chat_history.append(HumanMessage(content=human_message))
    ai_message=llm.invoke(chat_history)
    print(ai_message.content)
    chat_history.append(AIMessage(content=ai_message.content))
```
